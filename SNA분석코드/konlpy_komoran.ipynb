```python
# Count cooccurences
# input: nlp_file
# output: cooc_frq.txt, cooc_frq.graphml

# 하나의 네트워크 파일 만드는 수식 (동시출현)


import operator
#from cooc.my_graphml import MakeGraphml

def search_vocab(word):
    if w_dict.get(word):
        tid = w_dict.get(word)
        #print(word)
    else:
        w_dict[word] = len(w_dict) + 1  # {term: tid} 추가 
        tid = w_dict.get(word)
        t_dict[tid] = word          # {tid: term} 추가
    return tid


def count_cooc(aid, bid):
    if count.get((aid, bid)):
        count[aid,bid] = count[aid,bid] + 1
    else:
        count[aid,bid] = count[aid,bid] = 1
    #print(aid, bid, count[aid,bid])
    

''' main function '''
# input
input_file = 'LPG_YOUTUBE_2.txt'  ##형태소 분석 해줬던 파일 
# output: pair file
pair_file = 'cooc_LPG_YOUTUBE_2.txt'
# frq_threshold
thres = 1

f = open(input_file, 'r', encoding='utf-8')
out = open(pair_file, 'w', encoding='utf-8')

w_dict = {}   # dic {term: tid}
t_dict = {}   # dic {tid: term}
count = {}   # dic {(a,b):frq}

line = 0
for text in f.readlines():
    words = text.split('\t')[0].strip().split()   # 단어가 들어있는 칼럼[1]. 한줄씩 읽어와서 단어별로 분리
    line += 1
    #print(line)

    for i, a in enumerate(words):
        #print(i, a)
        aid = search_vocab(a)            # check tid
        
        for b in words[i+1:]:
            bid = search_vocab(b)        # check tid
            
            if aid == bid: continue     # 동일단어 pass
            elif aid < bid:
                count_cooc(aid, bid)
            else:                       # when a > b, 자리바꿈
                count_cooc(bid, aid)

    #break

#print(w_dict)
#print(t_dict)

# itemgetter(0): key 또는 itemgetter(1): value로 sort key, reverse=True (descending order)
sorted_value = sorted(count.items(), key=operator.itemgetter(1), reverse=True)
for i in range(len(sorted_value)):
    if sorted_value[i][1] >= thres:
        #print(sorted_value[i])
        print(t_dict[sorted_value[i][0][0]], t_dict[sorted_value[i][0][1]], sorted_value[i][1])
        out.write('{}\t{}\t{}\n'.format(t_dict[sorted_value[i][0][0]], t_dict[sorted_value[i][0][1]], sorted_value[i][1]))

print("===== Done =====")
f.close()
out.close()
```

```python
# my_graphml.py
# option: graphml file
# gephi 네트워크 만들어주기


pair_file = 'cooc_LPG_YOUTUBE_2.txt'
graphml_file = 'cooc_LPG_YOUTUBE_2.graphml'

pairs = open(pair_file, 'r', encoding = 'utf-8')
out = open(graphml_file, 'w', encoding = 'utf-8')

entity = []
e_dict = {}
count = []
for pair in pairs.readlines():
    e1 = pair.split('\t')[0]
    e2 = pair.split('\t')[1]
    #frq = ((word_dict[e1], word_dict[e2]),  pair.split('\t')[2])
    frq = ((e1, e2), pair.split('\t')[2])
    if frq not in count: count.append(frq)   # ((a, b), frq)
    if e1 not in entity: entity.append(e1)
    if e2 not in entity: entity.append(e2)
print('# terms: %s'% len(entity))

#create e_dict {entity: id} from entity
for i, w in enumerate(entity):
    e_dict[w] = i + 1 # {word: id}

out.write(
    "<?xml version=\"1.0\" encoding=\"UTF-8\"?><graphml xmlns=\"http://graphml.graphdrawing.org/xmlns\" xmlns:xsi=\"http://www.w3.org/2001/XMLSchema-instance\" xsi:schemaLocation=\"http://graphml.graphdrawing.org/xmlnshttp://graphml.graphdrawing.org/xmlns/1.0/graphml.xsd\">" +
    "<key id=\"d1\" for=\"edge\" attr.name=\"weight\" attr.type=\"double\"/>" +
    "<key id=\"d0\" for=\"node\" attr.name=\"label\" attr.type=\"string\"/>" +
    "<graph id=\"Entity\" edgedefault=\"undirected\">" + "\n")

# nodes
for i in entity:
    out.write("<node id=\"" + str(e_dict[i]) +"\">" + "\n")
    out.write("<data key=\"d0\">" + i + "</data>" + "\n")
    out.write("</node>")

# edges
for y in range(len(count)):
    out.write("<edge source=\"" + str(e_dict[count[y][0][0]]) + "\" target=\"" + str(e_dict[count[y][0][1]]) + "\">" + "\n")
    out.write("<data key=\"d1\">" + str(count[y][1]) + "</data>" + "\n")

    #out.write("<edge source=\"" + str(count[y][0][0]) + "\" target=\"" + str(count[y][0][1]) +"\">"+"\n")
    #out.write("<data key=\"d1\">" + str(count[y][1]) +"</data>"+"\n")
    out.write("</edge>")

out.write("</graph> </graphml>")
print('now you can see %s' % graphml_file)

pairs.close()
out.close()

# 1264 -> node가 1264개란 의미....
```
